### Inferential Statistics
Inferential Statistics analysis infers properties of a population. For example, by testing hypotheses and deriving estimates.

```python
pop3 = np.random.randint(10, 20, 1000)
np.random.seed(10)
estimates = [] # Create an empty list to hold point estimates

for x in range(200):
    sam3 = np.random.choice(a = pop3, size=100)
    estimates.append(sam3.mean())
```

```python
np.mean(pop3)
```

```python
pd.DataFrame(estimates).plot(kind="density")
```

### Point of Estimations
#### Confidence Interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
#### Margin of Error
margin_of_error = (critical_value) + (standard_deviation)/square_root(sample_size)

```python
import scipy.stats as stats

z_critical = stats.norm.ppf(q = 0.975) # Percent point function
z_critical
```

```python
t_critical = stats.t.ppf(q = 0.975, df = 24)
# 'df' is the Degrees of Freedom, which is found: sample size - 1 = degrees of freedom
t_critical
```

```python
# Calculate the margin of error
margin_of_error = z_critical * (np.std(estimates)/np.sqrt(200))
margin_of_error
```

```python
# Lower limit: sample_mean - margin_of_error
np.mean(estimates) - margin_of_error
```

```python
# Upper limit: sample_mean + margin_of_error
np.mean(estimates) + margin_of_error
```

#### Normal Distribution

<p>1 std dev = the middle 68.2% of the data

<p>3 std dev = the middle 99.7% of the data

```python
import matplotlib.pyplot as plt
import numpy as np

mu = 0.5
sigma = 0.1
s = np.random.normal(mu, sigma, 1000)

# Create bins and histogram
count, bins, ignored = plt.hist(s, 20, density=True)
# The video said normed = True, but that threw an error. Density = true works...

# Plot the distribution curve
plt.plot(bins, 1/(sigma*np.sqrt(2*np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='y')
plt.show()
```

## Tutorial #8
#### Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline

# Population size of 10000
population = np.random.binomial(10,0.4,10000)

sample_means=[]
for x in range(1,21):
    sample = np.random.choice(population, size = int(0.30 * len(population)), replace=False)
    sample_means.append(np.mean(sample))
    
# Calculating mean of sample means and population mean
mean_of_sample_means = np.mean(sample_means)
population_mean = np.mean(population)

print(f'Mean of Sample Means {mean_of_sample_means}')
print(f'Mean of Population {population_mean}')
```

```python
ax = sns.distplot(sample_means, kde=True, color='orange', hist_kws={"linewidth": 15, 'alpha':1})
```
